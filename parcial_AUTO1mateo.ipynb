F. Desarrolle un subplot con dos gráficos, el primero es un boxplot relacione artist_count
(eje x) con streams, y el segundo un histograma de los años de lanzamiento.

```python
tec_col={} #diccionario para 
for i in df['streams'].unique(): #creamos un for que itere a lo largo de cada tipo de ChestPainType
  tec_col[i]=df[df['streams']==i].artist_count #asignamos al diccionario cada tipo de dolor como clave y los valores de colesterol asociados
###print (f'Analizando el tipo de dolor {i}, los datos de colesterol son: \n {df[df['ChestPainType']==i].Cholesterol}')
print(tec_col)
```

```python
plt.boxplot([x for x in tec_col.values()], labels=[x for x in tec_col.keys()])
plt.xlabel('artist_conut ')
plt.ylabel('streams')
plt.title('pnuto F')
plt.show()
```
